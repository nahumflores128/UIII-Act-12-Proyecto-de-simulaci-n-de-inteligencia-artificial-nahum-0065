```python
# ==========================================================
# SISTEMA IA - GESTIÓN DE ADMINISTRADORES
# CBTIS 128
# Carrera: Inteligencia Artificial
# Materia: Elabora Proyectos de Programación Lógica
# Docente: Eliseo Nava
# Alumno: Nahum Flores
# NL = 21
# ==========================================================

class administrador:

    # Constructor
    def __init__(self):
        self.lista_administradores = []

    # Función para agregar administrador
    def agregar_administrador(self):

        print("\n===== REGISTRO DE ADMINISTRADOR =====")

        empresa = input("Empresa: ")
        departamento = input("Departamento: ")
        presupuesto = input("Presupuesto asignado: ")
        personal = input("Personal a cargo: ")
        metas = input("Metas trimestrales: ")
        software = input("Software ERP: ")

        datos = {
            "empresa": empresa,
            "departamento": departamento,
            "presupuesto": presupuesto,
            "personal": personal,
            "metas": metas,
            "software": software
        }

        self.lista_administradores.append(datos)

        print("\n✅ Administrador registrado correctamente.")

    # Función para listar administradores
    def listar_administradores(self):

        print("\n================ LISTA DE ADMINISTRADORES ================")

        if len(self.lista_administradores) == 0:
            print("No hay administradores registrados.")
        else:
            # Calculate maximum width for each column
            max_empresa = max([len(admin['empresa']) for admin in self.lista_administradores] + [len('EMPRESA')])
            max_departamento = max([len(admin['departamento']) for admin in self.lista_administradores] + [len('DEPARTAMENTO')])
            max_presupuesto = max([len(admin['presupuesto']) for admin in self.lista_administradores] + [len('PRESUPUESTO')])
            max_personal = max([len(admin['personal']) for admin in self.lista_administradores] + [len('PERSONAL')])
            max_metas = max([len(admin['metas']) for admin in self.lista_administradores] + [len('METAS')])
            max_software = max([len(admin['software']) for admin in self.lista_administradores] + [len('SOFTWARE ERP')])

            # Add some padding
            padding = 3
            max_empresa += padding
            max_departamento += padding
            max_presupuesto += padding
            max_personal += padding
            max_metas += padding
            max_software += padding

            header_format = (
                f"{'No.':<5}"
                f"{'EMPRESA':<{max_empresa}}"
                f"{'DEPARTAMENTO':<{max_departamento}}"
                f"{'PRESUPUESTO':<{max_presupuesto}}"
                f"{'PERSONAL':<{max_personal}}"
                f"{'METAS':<{max_metas}}"
                f"{'SOFTWARE ERP':<{max_software}}"
            )
            print(header_format)

            print("-" * (5 + max_empresa + max_departamento + max_presupuesto + max_personal + max_metas + max_software))

            contador = 1

            for admin in self.lista_administradores:

                row_format = (
                    f"{contador:<5}"
                    f"{admin['empresa']:<{max_empresa}}"
                    f"{admin['departamento']:<{max_departamento}}"
                    f"{admin['presupuesto']:<{max_presupuesto}}"
                    f"{admin['personal']:<{max_personal}}"
                    f"{admin['metas']:<{max_metas}}"
                    f"{admin['software']:<{max_software}}"
                )
                print(row_format)

                contador += 1

    # Función para actualizar administrador
    def actualizar_administrador(self):

        self.listar_administradores()

        if len(self.lista_administradores) > 0:
            try:
                numero = int(input("\nNúmero del administrador a actualizar: "))

                if numero >= 1 and numero <= len(self.lista_administradores):

                    print("\n===== NUEVOS DATOS =====")

                    self.lista_administradores[numero - 1]["empresa"] = input("Nueva empresa: ")
                    self.lista_administradores[numero - 1]["departamento"] = input("Nuevo departamento: ")
                    self.lista_administradores[numero - 1]["presupuesto"] = input("Nuevo presupuesto: ")
                    self.lista_administradores[numero - 1]["personal"] = input("Nuevo personal a cargo: ")
                    self.lista_administradores[numero - 1]["metas"] = input("Nuevas metas trimestrales: ")
                    self.lista_administradores[numero - 1]["software"] = input("Nuevo software ERP: ")

                    print("\n✅ Administrador actualizado correctamente.")

                else:
                    print("❌ Número incorrecto. Por favor, introduce un número dentro del rango mostrado.")
            except ValueError:
                print("❌ Entrada inválida. Por favor, introduce un número entero.")

    # Función para borrar administrador
    def borrar_administrador(self):

        self.listar_administradores()

        if len(self.lista_administradores) > 0:
            try:
                numero = int(input("\nNúmero del administrador a borrar: "))

                if numero >= 1 and numero <= len(self.lista_administradores):

                    self.lista_administradores.pop(numero - 1)

                    print("\n🗑️ Administrador eliminado correctamente.")

                else:
                    print("❌ Número incorrecto. Por favor, introduce un número dentro del rango mostrado.")
            except ValueError:
                print("❌ Entrada inválida. Por favor, introduce un número entero.")


# ==========================================================
# PROGRAMA PRINCIPAL
# ==========================================================

sistema = administrador()

opcion = 0

while opcion != 5:

    print("\n======================================")
    print("   IA DE GESTIÓN DE ADMINISTRADORES")
    print("======================================")
    print("1. Agregar administrador")
    print("2. Listar administradores")
    print("3. Actualizar administrador")
    print("4. Borrar administrador")
    print("5. Salir")
    print("======================================")

    opcion = int(input("Seleccione una opción: "))

    if opcion == 1:
        sistema.agregar_administrador()

    elif opcion == 2:
        sistema.listar_administradores()

    elif opcion == 3:
        sistema.actualizar_administrador()

    elif opcion == 4:
        sistema.borrar_administrador()

    elif opcion == 5:
        print("\n👋 Saliendo del sistema...")
        print("Gracias por utilizar el sistema IA.")

    else:
        print("\n❌ Opción inválida.")
```
